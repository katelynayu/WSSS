```python
def svp(T):
    L = 2.5 * math.pow(10,6) #J/kg
    R_v = 461 #J /(kg K)
    return .611 * math.exp((L/R_v)*((1/273)-(1/(T)))) #saturation vapor pressure #kPa 

svp1 = []
svp2 = []
svp3 = []
svp4 = []
svp5 = []
svp6 = []
svp7 = []
svp8 = []
svp9 = []
svp10 = []

def svplist(time):
    svp1.append((svp(temp1_CS215[time]+273.15)) * 760/101.325)
    svp2.append((svp(temp2_CS215[time]+273.15))* 760/101.325)
    svp3.append((svp(temp3_CS215[time]+273.15))* 760/101.325)
    svp4.append((svp(temp4_CS215[time]+273.15))* 760/101.325)
    svp5.append((svp(temp5_CS215[time]+273.15))* 760/101.325)
    svp6.append((svp(temp6_CS215[time]+273.15))* 760/101.325)
    svp7.append((svp(temp7_CS215[time]+273.15))* 760/101.325)
    svp8.append((svp(float(temp8_CS215[time])+273.15))* 760/101.325)
    svp9.append((svp(temp9_CS215[time]+273.15))* 760/101.325)
    svp10.append((svp(temp10_CS215[time]+273.15))* 760/101.325) 

for i in range(0, 30859):
    svplist(i)
print(svp1)

RH1 = []
RH2 = []
RH3 = []
RH4 = []
RH5 = []
RH6 = []
RH7 = []
RH8 = []
RH9 = []
RH10 = []

def RH(time):
    RH1.append(RH1_CS215[time] /100)
    RH2.append(RH2_CS215[time]/100)
    RH3.append(RH3_CS215[time]/100)
    RH4.append(RH4_CS215[time]/100)
    RH5.append(RH5_CS215[time]/100)
    RH6.append(RH6_CS215[time]/100)
    RH7.append(RH7_CS215[time]/100)
    RH8.append(float(RH8_CS215[time])/100)
    RH9.append(RH9_CS215[time]/100)
    RH10.append(RH10_CS215[time]/100)

for i in range(0, 30859):
    RH(i)
print(RH1)
```

```python
ad1 = []
ad2 = []
ad3 = []
ad4 = []
ad5 = []
ad6 = []
ad7 = []
ad8 = []
ad9 = []
ad10 = []

def airdensity(pressure, temperature, pp_w):
    #ideal gas equation: rho_humid = (rho_d * M_d + rho_v * M_v) / RT
    P = np.multiply(133.3,pressure)
    R = 8.314 #J/(KÂ·mol)
    rho_d = P - pp_w #
    M_d = 0.028964 #kg/mol
    rho_v = np.multiply(133.3, pp_w)
    M_v = 0.018016 #kg/mol
    T = temperature #K
    return ((np.multiply(rho_d, M_d) + np.multiply(rho_v, M_v))/ np.multiply(R , T))

for i in range(0, 30859):
    ad1.append(airdensity(P1[i], T1[i], pp1[i]))
    ad2.append(airdensity(P2[i], T2[i], pp2[i]))
    ad3.append(airdensity(P3[i], T3[i], pp3[i]))
    ad4.append(airdensity(P4[i], T4[i], pp4[i]))
    ad5.append(airdensity(P5[i], T5[i], pp5[i]))
    ad6.append(airdensity(P6[i], T6[i], pp6[i]))
    ad7.append(airdensity(P7[i], T7[i], pp7[i]))
    ad8.append(airdensity(P8[i], T8[i], pp8[i]))
    ad9.append(airdensity(P9[i], T9[i], pp9[i]))
    ad1.append(airdensity(P10[i], T10[i], pp10[i]))

height = [7.9, 10, 12.1, 14.2, 16.3, 18.4, 20.5, 22.6, 24.7, 26.8]

print(ad1)
```
